```python
#Accesing API Key
from google.colab import userdata
sec_key=userdata.get('Huggingface')
print(sec_key)
```

```python
from langchain_huggingface import HuggingFaceEndpoint
import os
os.environ["HUGGINGFACEHUB_API_TOKEN"] = sec_key


```

```python
#Accesing the Model using an API
repo_id="mistralai/Mistral-7B-Instruct-v0.3"
llm=HuggingFaceEndpoint(repo_id=repo_id,max_length=128,temperature=0.7,token=sec_key)
```

```python
llm.invoke("What is a neural network?")
```

```python
#Introducing PromptTemplates and LLMChains
from langchain_core.prompts import PromptTemplate
from langchain.chains import LLMChain
```

```python
question="Who is the CEO of HuggingFace? "
template="Question:{question}\nAnswer:"
prompt=PromptTemplate(template=template,input_variables=["question"])
print(prompt)
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
print(llm_chain.run(question))
```

```python

#Accesing Models using HuggingFacePipeline
from langchain_huggingface import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer,pipeline
```

```python
model_id="openai-community/gpt2"
model=AutoModelForCausalLM.from_pretrained(model_id)
tokenizer=AutoTokenizer.from_pretrained(model_id)
```

```python
pipeline=pipeline("text-generation",model=model,tokenizer=tokenizer,max_new_tokens=100)
hf=HuggingFacePipeline(pipeline=pipeline)
```

```python
hf
```

```python
hf.invoke("Langchain is a company")
```

```python
hf.invoke("Aliens are real")
```

```python
#LLMS Running on GPU
gpu_llm=HuggingFacePipeline.from_model_id(model_id=model_id,task="text-generation",device=0)
chain=LLMChain(llm=gpu_llm,prompt=prompt)
print(chain.run(question))
```
